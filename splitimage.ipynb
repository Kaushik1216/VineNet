```python
import os

def get_filenames(folder_path):
  
  filenames = []
  # Check if the path is a directory
  if os.path.isdir(folder_path):
    # Get all entries in the directory
    for entry in os.listdir(folder_path):
      # Check if the entry is a file (not a directory)
      if os.path.isfile(os.path.join(folder_path, entry)):
        filenames.append(entry)
  else:
    print(f"Error: '{folder_path}' is not a valid directory.")

  return filenames

# Example usage
folder_path1 = "./dataset/images/val"
filenames1 = get_filenames(folder_path1)

print("no of images 1  : " , len(filenames1) , len(set(filenames1)))

folder_path2 = "./dataset/masks/val"
filenames2 = get_filenames(folder_path2)

print("no of images 2  : " , len(filenames2) , len(set(filenames2)))

```

```python
def random_select_images(image_list, num_images):
  
  if num_images > len(image_list):
    raise ValueError("Cannot select more images than available. Requested:", num_images, ", Available:", len(image_list))

  # Create a copy to avoid modifying the original list
  selected_images = list(image_list)

  # Randomly shuffle the list to ensure unbiased selection
  random.shuffle(selected_images)

  return selected_images[:num_images]
```

```python
import os
import shutil

def copy_and_delete_files(source_path, destination_path, filenames):
  """
  This function copies files from a source folder to a destination folder,
  and deletes the copied files from the source folder. It uses a list of filenames
  to determine which files to copy.

  Args:
      source_path (str): The path to the source folder.
      destination_path (str): The path to the destination folder.
      filenames (list): A list of filenames to copy.
  """

  # Check if source and destination paths are valid directories
  if not os.path.isdir(source_path):
    print(f"Error: '{source_path}' is not a valid source directory.")
    return
  if not os.path.isdir(destination_path):
    print(f"Error: '{destination_path}' is not a valid destination directory.")
    return

  # Iterate through the list of filenames
  for filename in filenames:
    # Construct full paths for source and destination files
    source_file = os.path.join(source_path, filename)
    destination_file = os.path.join(destination_path, filename)

    # Check if the file exists in the source folder
    if not os.path.isfile(source_file):
      print(f"File not found: {filename}")
      continue

    # Copy the file to the destination folder
    try:
      shutil.copy2(source_file, destination_file)  # Preserves file metadata
      print(f"Copied: {filename} -> {destination_path}")
      # Delete the copied file from the source folder
      os.remove(source_file)
      print(f"Deleted: {filename} from source")
    except OSError as e:
      print(f"Error copying/deleting {filename}: {e}")

# Example usage
source_path = "./dataset/images"  # Replace with your actual source folder path
destination_path = "./dataset/images/val"  # Replace with your actual destination folder path
# filenames_to_copy = ["image1.jpg", "data.txt"]  # Replace with filenames to copy

# copy_and_delete_files(source_path, destination_path, listfilter)
```
